```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
scan_points = pd.read_csv('custom_scan.csv', names=['freq'])
data_points = pd.read_csv('n30_unpolarized_dimer.csv', names=['freq','counts'])
```

```python
plt.figure()
# Plotting vertical lines to visually observe data point densities
for xc in scan_points.values:
    plt.axvline(x=xc, linewidth=0.1)

plt.semilogy(data_points['freq'],data_points['counts'])

plt.xlim(min(scan_points.values), max(scan_points.values))
plt.ylim(1, 1000)
```
